```python
## Cropping (You will need this in project 1)
emir = plt.imread("emir.tif")
height, width = emir.shape
print(height, width)
plt.imshow(emir)
plt.show()
```

```python
emir1 = emir[:int(height/3)]
emir2 = emir[int(height/3):int(2*height/3)]
emir3 = emir[int(2*height/3):]
z = np.stack(np.asarray([emir1, emir2, emir3]), axis=2)
plt.imshow(emir1)
emir2
```

```python
## Resizing
size = (np.array(emir.shape) * 3.5).astype(int)
res_im = sktr.resize(emir, size, order=3)
plt.imshow(res_im)
print('Original image size: %d by %d' % (emir.shape[0], emir.shape[1]))
print('Resized image size: %d by %d' % (res_im.shape[0], res_im.shape[1]))
```

```python
## Averaging Photos
import glob
image_list = glob.glob('./starfish/*.jpg')
print('number of images = %d' % len(image_list))
# load and display first 10 images
num_imgs = 10
for n in range(num_imgs): 
    im = plt.imread(image_list[n])
    plt.imshow(im)
    plt.show()
```

```python
%%time
# compute an average image using FOR loops

average_im = np.zeros((256, 256, 3), np.float64)

# sum over all the images
for image_path in image_list:
    im = plt.imread(image_path)
    im_edit = (im*255.0)/im.max()
    im_edit = sktr.resize(im, [256, 256])
    for x in range(256):
        for y in range(256):
            for c in range(3):
                average_im[x, y, c] += im_edit[x, y, c]
                
# compute the average
average_im = average_im / len(image_list)
plt.imshow(average_im)
plt.show()
```

```python
%%time

average_im = np.zeros((256, 256, 3), np.float64)
# sum over all the images
for image_path in image_list:
    im = plt.imread(image_path)
    im_edit = (im*255.0)/im.max()
    im_edit = sktr.resize(im, [256, 256])
    average_im = average_im + im_edit
    
# compute the average 
average_im = average_im / len(image_list)
plt.imshow(average_im)
```
