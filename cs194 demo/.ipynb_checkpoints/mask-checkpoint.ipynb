```python
# Load and display the image
input_file = 'fruit.jpg'
im = imread(input_file) / 255.0
imshow(im)
```

```python
# Find pixel locations where colors match background (white)
bg = [1.0, 1.0, 1.0]
thresh = 0.10
mask = (abs(im[:,:,0] - bg[0]) < thresh) & \
       (abs(im[:,:,1] - bg[1]) < thresh) & \
       (abs(im[:,:,2] - bg[2]) < thresh)
imshow(mask, cmap='gray')
```

```python
# Replace background color with green
new_bg = [0.0, 1.0, 0]
masked_im = np.copy(im)
masked_im[mask, 0] = new_bg[0]
masked_im[mask, 1] = new_bg[1]
masked_im[mask, 2] = new_bg[2]
imshow(masked_im)
```
